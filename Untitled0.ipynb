```python
# Import Spark session to work with PySpark
from pyspark.sql import SparkSession

# Import functions for analysis
from pyspark.sql.functions import avg, count, col

# For visualization
import matplotlib.pyplot as plt
```

```python
# Create a Spark session (entry point for PySpark)
spark = SparkSession.builder \
    .appName("FlightDelayAnalysis") \
    .getOrCreate()
```

```python
# Path where your CSV file is uploaded in Colab
file_path = "/content/flight_data_2024.csv"
```

```python
import os

print(os.listdir('/content'))
```

```python
df = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .option("sep", ",") \
    .csv("/content/flight_data_2024.csv")
```

```python
# Load CSV file into Spark DataFrame
df = spark.read.csv(file_path, header=True, inferSchema=True)

# Display first 5 rows
df.show(5)
```

```python
# Print structure (column names and data types)
df.printSchema()

# Display all column names
print(df.columns)

# Show statistical summary
df.describe().show()
```

```python
# Fill missing values instead of dropping everything
df = df.fillna(0)
```

```python
df.columns
```

```python
# Group data by airline and count number of flights
df.groupBy("op_unique_carrier") \
  .count() \
  .orderBy("count", ascending=False) \
  .show()
```

```python
# Correct column: arr_delay
from pyspark.sql.functions import avg

df.groupBy("op_unique_carrier") \
  .agg(avg("arr_delay").alias("avg_arrival_delay")) \
  .orderBy("avg_arrival_delay", ascending=False) \
  .show()
```

```python
# Correct column: month
df.groupBy("month") \
  .count() \
  .orderBy("month") \
  .show()
```

```python
# Correct columns: origin, dest, arr_delay
df.groupBy("origin", "dest") \
  .agg(avg("arr_delay").alias("avg_delay")) \
  .orderBy("avg_delay", ascending=False) \
  .show(10)
```

```python
df.orderBy(col("dep_delay").desc()).show(10)
```

```python
print("Partitions before:", df.rdd.getNumPartitions())

df = df.repartition(4)

print("Partitions after:", df.rdd.getNumPartitions())
```

```python
df.cache()
```

```python
# Check row count before converting
print("Row count:", df.count())

# To avoid out-of-memory errors with large datasets, convert a limited number of rows to Pandas.
# This ensures pandas_df is not empty and allows for local exploration without crashing.
pandas_df = df.limit(1000).toPandas()
print("Shape of limited Pandas DataFrame:", pandas_df.shape)
```

```python
if pandas_df.shape[0] > 0:
    pandas_df['op_unique_carrier'].value_counts().head(10).plot(kind='bar')

    plt.title("Top Airlines by Flights")
    plt.xlabel("Airline")
    plt.ylabel("Number of Flights")

    plt.show()
else:
    print("No data available for plotting")
```

```python
if pandas_df.shape[0] > 0:
    pandas_df['month'].value_counts().sort_index().plot(kind='line')

    plt.title("Flights per Month")
    plt.xlabel("Month")
    plt.ylabel("Number of Flights")

    plt.show()
```
